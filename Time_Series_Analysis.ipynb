```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('portfolio_data.csv')
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'] + ' 00:00:00')
df.set_index('Date',inplace=True)
```

```python
resample_df = df.resample('D').mean()
```

```python
resample_df.plot(figsize=(10,6))
plt.title("Time Analysis Date")
plt.xlabel("Data")
plt.ylabel("Value")
plt.legend(resample_df.columns)
plt.show
```
